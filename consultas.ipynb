```python
## Qual a quantidade usuários que moram em "United States" gostam de Windows?
session = Session()
result = session.execute(
    " SELECT p.nome, so.nome as sist_o, COUNT(r.index ) as qtd_usuarios "\
    " FROM respondente r "\
    " INNER JOIN sistema_operacional so ON r.sistema_operacional_index = so.index "\
    " INNER JOIN pais p ON p.index = r.pais_index  "\
    " WHERE p.nome = 'United States' and so.nome = 'Windows' "
    " GROUP BY p.nome, so.nome  "\
)
print('Quantidade de usuários que moram em "United States" e que gostam de Windows')
for row in result:
    print(row)

```

```python
## Qual a média e o desvio padrão do salário dos usuários que usam Slack para cada tamanho de empresa disponível?
session = Session()
result = session.execute(
    " SELECT e.tamanho, avg(r.salario) as media_salarios, stddev(r.salario) as desvio_padrao "\
    " FROM respondente r "\
    " INNER JOIN respondenteferramenta_comunicacao rc ON r.index = rc.respondente_index "\
    " INNER JOIN ferramenta_comunicacao fc ON rc.ferramenta_comunicacao_index = fc.index "\
    " INNER JOIN empresa e ON r.empresa_index = e.index "\
    " WHERE fc.nome = 'Slack' "\
    " GROUP BY fc.nome, e.tamanho  "\
)
print('Média e o desvio padrão do salário dos usuários que usam Slack para cada tamanho de empresa disponível')
for row in result:
    print(row)
```
